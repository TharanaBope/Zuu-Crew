```python
df = pd.read_csv(r'E:\Courses and Certifications\Zuu Crew\Building Production Ready ML Systems\Week 01\Data\Processed\ChurnModelling_Outliers_Handeled.csv')
df.head()
```

350 - 580 => 'Poor'
580 - 670 => 'Fair'
670 - 740 => 'Good'
740 - 800 => 'Very Good'
800 - 850 => 'Execellent'

```python
def custom_binning_credit_score(score):
    if score < 580:
        return 'Poor'
    elif score < 670:
        return 'Fair'
    elif score < 740:
        return 'Good'
    elif score < 800:
        return 'Very Good'
    elif score < 850:
        return 'Excellent'
    else:
        return 'Exceptional'  # Handle scores >= 850 (like the 850 in row 4)

# Apply the binning
df['CreditScoreBins'] = df['CreditScore'].apply(custom_binning_credit_score)

# Drop the original CreditScore column
df = df.drop('CreditScore', axis=1)

# Display the result
df.head(10)
```
