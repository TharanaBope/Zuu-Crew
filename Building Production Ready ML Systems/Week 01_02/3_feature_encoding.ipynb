| Variable Type | Preferred Encoding | Why? |
|--------------|-------------------|------|
| Nominal | One-Hot Encoding | No inherent order → avoids implying false ordinal relationships |
| Ordinal | Label Encoding | Preserves order → small integers represent increasing levels |

Gender <Male, Female>

        Gender_Male, Gender_Female
Male    [1            0]
Female  [0            1]

(Assume Gender is Ordinal);
    Male -> 1
    Female -> 0

Gender -> Nominal
Geography -> Nominal
CreditScoreBins -> Ordinal

```python
df = pd.read_csv(r'E:\Courses and Certifications\0 Zuu Crew\Building Production Ready ML Systems\Week 01_02\Data\Processed\ChurnModelling_Binning_Applied.csv')
df.head()
```

### 2. Encode Norminal variable

```python
norminal_variables = ['Geography', 'Gender']

geography_dummies = pd.get_dummies(df['Geography'], prefix = "Geography")
gender_dummies = pd.get_dummies(df['Gender'], prefix = "Gender")

df_encoded = pd.concat([df, geography_dummies], axis=1) #axis = 1 means concatenate along columns
del df_encoded['Geography']

df_encoded = pd.concat([df_encoded, gender_dummies], axis=1)
del df_encoded['Gender']

df_encoded
```

### 3. Encode Ordinal variable

```python
encode_dict_creditscore = {'Poor': 0,
                           'Fair': 1,
                           'Tenure':2,
                           'Very Good':3,
                           'Excelent':4
                           }
df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_creditscore)
df_encoded.head(10)
```

```python
df_encoded.to_csv(r'E:\Courses and Certifications\0 Zuu Crew\Building Production Ready ML Systems\Week 01_02\Data\Processed\ChurnModelling_Encoded.csv', index=False)
```
